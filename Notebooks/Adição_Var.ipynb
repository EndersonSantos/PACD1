## WEB SCRAPING

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Carrega o arquivo "links.csv"
df_links = pd.read_csv('link.csv')

# Lista para armazenar as informações extraídas
data = []

# Loop pelos links
for link in df_links['LinkPlayer']:
    # Faz a requisição HTTP
    response = requests.get(link)
    
    # Verifica se a requisição foi bem sucedida
    if response.status_code == 200:
        # Extrai o conteúdo HTML
        html = response.content
        
        # Cria o objeto BeautifulSoup
        soup = BeautifulSoup(html, 'html.parser')
        
        # Extrai o país
        country = soup.find('div', {'class': 'player-flag-code'}).text.strip()
        
        # Extrai o nome
        name = soup.find('title').text.split('|')[0].strip()
        
        # Adiciona as informações à lista
        data.append({'Country': country, 'Name': name})
        
# Cria um DataFrame com as informações extraídas
df_info = pd.DataFrame(data)

# Exibe o resultado
print(df_info.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# carrega o DataFrame com os links dos jogadores
df_links = pd.read_csv('link.csv')

resultados = []  # lista para armazenar os resultados

for link in df_links['LinkPlayer']:
    # faz o request da página
    response = requests.get(link)
    # cria o objeto BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    # encontra o elemento HTML com a tag <meta> e o atributo name="twitter:title"
    nome_element = soup.find('meta', {'name': 'twitter:title'})
    # extrai o nome do jogador do atributo 'content' do elemento HTML
    nome = nome_element('content')
    # adiciona o resultado na lista de resultados
    resultados.append({'link': link, 'nome': nome})

# cria um novo DataFrame com os resultados
df_nomes = pd.DataFrame(resultados)

# imprime o DataFrame com os nomes dos jogadores
print(df_nomes)
```

```python
# importar as bibliotecas necessárias
import pandas as pd
import requests
from bs4 import BeautifulSoup

# ler o ficheiro csv com os links
df_links = pd.read_csv('link.csv')

# criar uma lista para guardar os resultados
resultados = []

# percorrer todos os links e extrair a informação
for link in df_links['LinkPlayer']:
    # fazer o request à página e guardar o conteúdo HTML
    page = requests.get(link)
    html = page.content
    
    # criar um objeto BeautifulSoup para facilitar a extração de informação do HTML
    soup = BeautifulSoup(html, 'html.parser')
    
    # encontrar o elemento HTML com a flag code do país
    pais_element = soup.find('div', {'class': 'player-flag'})
    
    # extrair a flag code do atributo 'class' do elemento HTML
    if pais_element is not None:
        pais = pais_element.get('class')[1]
    else:
        pais = None
    
    # adicionar o resultado na lista de resultados
    resultados.append({'link': link, 'pais': pais})

# criar um novo dataframe com os resultados e guardar em csv
df_paises = pd.DataFrame(resultados)
df_paises.to_csv('paises.csv', index=False)


```
