## WEB SCRAPING

```python
df1 = pd.read_csv('atp_players.csv')
df1
```

```python
df1['full_name'] = df1['name_first']  + ' ' + df1['name_last']
```

```python
df1 
```

```python
# Crie um dicionário que mapeia nomes de jogadores para códigos de país
jogador_pais_map = dict(zip(df1['full_name'], df1['ioc']))

# Use o método 'map' para mapear os códigos de país na coluna 'Player1_flag'
df_merged1['Player1_flag'] = df_merged1['Player1_Name'].map(jogador_pais_map)
```

```python
# Crie um dicionário que mapeia nomes de jogadores para códigos de país
jogador_pais_map = dict(zip(df1['full_name'], df1['ioc']))

# Use o método 'map' para mapear os códigos de país na coluna 'Player1_flag'
df_merged1['Player2_flag'] = df_merged1['Player2_Name'].map(jogador_pais_map)
```

```python
df_merged1
```

```python
null_flags = df_merged1.loc[df_merged1['Player1_flag'].isnull()]
null_names = null_flags['Player1_Name'].unique()
print(null_names)
```

```python
null_flags = df_merged1.loc[df_merged1['Player2_flag'].isnull()]
null_names = null_flags['Player2_Name'].unique()
print(null_names)
```

```python
# agrupando os dados pela altura e calculando a média dos pesos por altura
media_pesos_por_altura = df_merged1.groupby('Player1_Height')['Player1_Weight'].mean()

# preenchendo os valores NaN da coluna 'peso' com a média dos pesos correspondentes
df_merged1['Player1_Weight'] = df_merged1['Player1_Weight'].fillna(df_merged1['Player1_Height'].map(media_pesos_por_altura))
```

```python
# agrupando os dados pela altura e calculando a média dos pesos por altura
media_pesos_por_altura = df_merged1.groupby('Player2_Height')['Player2_Weight'].mean()

# preenchendo os valores NaN da coluna 'peso' com a média dos pesos correspondentes
df_merged1['Player2_Weight'] = df_merged1['Player2_Weight'].fillna(df_merged1['Player2_Height'].map(media_pesos_por_altura))
```

```python
df_merged1[['Player1_Weight','Player2_Weight']] = df_merged1[['Player1_Weight','Player2_Weight']].round(0)
```
