```python
import time

start_time = time.time()


from lxml import etree
import requests
from bs4 import BeautifulSoup
import pandas as pd

def extract_element(url):
    headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36 OPR/98.0.0.0"}
    site = requests.get(url=url, headers=headers)
    soup = BeautifulSoup(site.text, "html.parser")
    dom = etree.HTML(str(soup))
    element = dom.xpath('//div[@class="player-flag-code"]/text()')
    return element[0] if element else ''

def extract_name(url):
    headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36 OPR/98.0.0.0"}
    site = requests.get(url=url, headers=headers)
    soup = BeautifulSoup(site.text, "html.parser")
    name = soup.find('meta', {'property': 'pageTransitionTitle'})['content']
    return name



# carrega o dataframe com os links
df_links = pd.read_csv('link.csv')

# extrai o elemento de cada URL e adiciona uma nova coluna no dataframe
df_links['player-flag-code'] = df_links['LinkPlayer'].apply(extract_element)
# adiciona uma nova coluna com o nome do jogador
df_links['player-name'] = df_links['LinkPlayer'].apply(extract_name)
# mostra o dataframe resultante
print(df_links)

end_time = time.time()
print("Tempo de execução: ", end_time - start_time, " segundos")
```

```python
from lxml import etree
import requests
from bs4 import BeautifulSoup
import pandas as pd

def extract_element(url):
    headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36 OPR/98.0.0.0"}
    site = requests.get(url=url, headers=headers)
    soup = BeautifulSoup(site.text, "html.parser")
    title_element = soup.find("h1", class_="page-title")
    name = title_element.text if title_element else ''
    dom = etree.HTML(str(soup))
    flag_element = dom.xpath('//div[@class="player-flag-code"]/text()')
    flag = flag_element[0] if flag_element else ''
    return name, flag

# carrega o dataframe com os links
df_links = pd.read_csv('link.csv')

# extrai o nome e o código do país de cada URL e adiciona novas colunas no dataframe
df_links[['player-name', 'player-flag-code']] = df_links['LinkPlayer'].apply(extract_element).apply(pd.Series)

# mostra o dataframe resultante
print(df_links)
```
